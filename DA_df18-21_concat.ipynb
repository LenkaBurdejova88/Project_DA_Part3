### Vertical concatenation of tables

```python
import pandas as pd
df18 = pd.read_csv("df18_all_col.csv")
df19 = pd.read_csv("df19_all_col.csv")
df20 = pd.read_csv("df20_all_col.csv")
df21 = pd.read_csv("df21_all_col.csv")

df_joined = pd.concat([df18, df19, df20, df21], ignore_index=True)

df_joined

df_joined.to_csv("df18-21.csv", index=False)

df_joined
```
